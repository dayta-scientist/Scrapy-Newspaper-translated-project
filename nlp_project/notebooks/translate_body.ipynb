```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/quotes.csv")
df
```

```python
title = df["url"].apply(lambda x: x.split('/')[-1])

title = title.apply(lambda x: x.replace('-', ' '))

title = title.apply(lambda x: x.capitalize())
```

```python
title
```

```python
df["title"] = title
df["title"]
```

```python
df.head(3)
```

```python
df["body"][0]
```

```python
df["body"] = df["body"].replace(to_replace ='(\.,)', value = '. ', regex = True) 
  
df.head()
```

```python
from googletrans import Translator, constants
# init the Google API translator
translator = Translator()
```

```python
df['English_body'] = df["body"].apply(translator.translate, src='es', dest='en').apply(getattr, args=('text',))
df
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["English_body"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["English_body"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["English_body"]]
df = df[["English_body", "Positive", "Negative", "Neutral"]]
df.head()
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk import word_tokenize
```

```python
df['tokens'] = df['English_body'].apply(lambda x: word_tokenize(x))
df.head()
```
